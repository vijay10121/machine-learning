```python
import pandas as pd
import numpy as np 
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=10
)

plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Classification Data")
plt.show()
```

```python
def gradient_descent(X,y,epochs,lr):
    X = np.insert(X,0,1,axis=1)
    W = np.ones(X.shape[1])

    for i in range(epochs):
        y_hat = sigmoid(np.dot(X,W))
        W = W + (lr/X.shape[0])*np.dot((y - y_hat),X)

    return W[1:],W[0]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
weights,intercept = gradient_descent(X,y,500,0.5)
```

```python
weights, intercept
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(X[:,0],-(np.dot(X[:,0],weights[0]+intercept))/weights[1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Classification Data")
plt.ylim(-3,3)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(solver='sag')
lor.fit(X,y)
```

```python
x_input1 = np.linspace(-3,3,100)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_input1,-(np.dot(x_input1,weights[0]+intercept))/weights[1])
plt.plot(x_input1,-(np.dot(x_input1,lor.coef_[0][0]+lor.intercept_[0]))/lor.coef_[0][1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Classification Data")
plt.ylim(-3,3)
plt.show()
```
